# The first notebook

**What you see here is a Google Colaboratory notebook.**

Using a notebook allows you to write and execute Python code in your browser. 

Each notebook consists of '**code cells**' in which you can edit and execute code or '**text cells**' which you can use to document the code (what you read is content of a code cell)
