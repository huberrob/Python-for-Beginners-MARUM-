## Creating and importing a module -foreword<br>
Modules refer to a file containing Python statements and definitions.
We can define our most used functions in a module and import it, instead of copying their definitions into different programs.<br>
There are 3 ways to import a file:
1. import somefile <br>
Everything in somefile.py gets imported.
To refer to something in the file, append the
text “somefile.” to the front of its name. 
2. from somefile import * <br>
Everything in somefile.py gets imported. To refer to anything in the module, just use its
name. Everything in the module is now in the
current namespace. 
3. from somefile import className <br>
Only the item className in somefile.py gets
imported.
After importing className, you can just use it
without a module prefix. It’s brought into the
current namespace.
Take care! Overwrites the definition of this
name if already defined in the current
namespace!
