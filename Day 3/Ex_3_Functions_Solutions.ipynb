Local variables can be accessed only inside the function in which they are declared, whereas global variables can be accessed throughout the program body by all functions.

1.   Call the sum function with arguments values, arg1 => 10 and arg2 => 20
2.   print the global variable 'total'





```
total = 0; # This is global variable.

def sum( arg1, arg2 ):
   # Add both the parameters and return them."
   total = arg1 + arg2; # Here total is local variable.
   print ("Inside the function local total : ", total)
   return total;
```

```
sum(10,20)
print('Global variable total :',total)
```

**Exercise 4**

The following function takes one argument x and returns:
1. 0 if x<=0
2. 5 * x if 0 < x <= 10
3. x^2 if x > 10 


```
import matplotlib.pyplot as plt
import numpy as np
```

```
def func(x):
    # branching
    if x<=0:
        return 0
    elif x <=10:
        return 5*x
    else:
        return x**2
```

```
# define variables
START=-10
STOP=20
STEP=0.1

# Use numpy.arange() function to create an array of x values.
arr=np.arange(START,STOP,STEP)
# calculate output array (results of application of func function)
func_output=[]

# calulate output array applying func(x) for each item in arr
for item in arr:
    # apply the function 'func'
    output_value=func(item)
    # append the output_value to the list 'func_output'
    func_output.append(output_value)
```

```
######################
# Plot this function using matplotlib.pyplot for x [-10,20]
######################

# increase the size of the plot
plt.figure(figsize=(10,10))
plt.plot(arr,func_output)
plt.grid()
# add information about axis
plt.xlabel("x",fontsize=18)
plt.ylabel("f(x)",fontsize=18)
```
